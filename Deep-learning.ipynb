In addition, the GELU activation fuction (the link for the [main paper](https://arxiv.org/pdf/1606.08415v5.pdf)) has been currently widely used in the state-of-the-art Vision for Transformers (e.g., here is the link for [the main ViT paper](https://arxiv.org/pdf/2010.11929v2.pdf)).

a) Answer
1. Output Range:
The ELU activation function is defined as:
$$
E L U(x)= \begin{cases}0.1(\exp (x)-1) & \text { if } x \leq 0 \\ x & \text { if } x>0\end{cases}
$$
- For $x \leq 0$ : The minimum value of $0.1(\exp (x)-1)$ is when $x$ approaches negative infinity, as exp $(x)$ become 0 when x is large which gives a value of -0.1 . So, the minimum value is -0.1 .
- For $x>0$ : The function is just $x$, so it can go to positive infinity. Therefore, the output range of ELU is $[-0.1, \infty)$.
2. Derivative:
Let's find the derivative of the ELU function:
For $x \leq 0$ :
$$
\frac{d}{d x}[0.1(\exp (x)-1)]=0.1 \exp (x)
$$
For $x>0$ :
$$
\frac{d}{d x}[x]=1
$$
So, the derivative of ELU is:
$$
E L U^{\prime}(x)= \begin{cases}0.1 \exp (x) & \text { if } x \leq 0 \\ 1 & \text { if } x>0\end{cases}
$$



```python
import matplotlib.pyplot as plt

def elu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=0.1):
    return np.where(x > 0, 1, alpha * np.exp(x))

x = np.linspace(-5, 5, 400)
y_elu = elu(x)
y_elu_derivative = elu_derivative(x)

plt.figure(figsize=(10, 6))

# Plot ELU
plt.plot(x, y_elu, label='ELU(x)', color='blue')

# Plot ELU Derivative
plt.plot(x, y_elu_derivative, label="ELU'(x)", color='red', linestyle='--')

plt.title('ELU Activation Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Answer g)

----->Why is the cross-entropy loss always non-negative?
The cross-entropy loss is defined as:
$$
C E(p, q)=-\sum_{i=1}^M p_i \log \left(q_i\right)
$$
where:
- $p$ is the true distribution (in our case, the one-hot encoded vector).
- $q$ is the predicted distribution (softmax probabilities from the neural network).
- $M$ is the number of classes.
Given that $p_i$ is either 0 or 1 (due to one-hot encoding), and $q_i$ is a probability value between 0 and 1 , the term $p_i \log \left(q_i\right)$ is always non-positive. Summing up these non-positive values (multiplied by -1 ) results in a non-negative value.

----->When does the cross-entropy loss get the value 0 ?
The cross-entropy loss will be 0 if and only if for every class $i$ :
- $p_i=1$ and $q_i=1$ OR
- $p_i=0$ and $q_i=0$
However, in practice, the output of a softmax function (which gives the probabilities $q$ ) will never be exactly 0 or 1 . But for the sake of understanding, if the neural network predicts the exact true class with absolute certainty (i.e., probability 1 for the true class and 0 for all others), then the cross-entropy loss will be 0 .
For a general feed-forward neural network with $M=4$ classes:
- If $y$ is the true class, then the one-hot encoded vector $p$ will have a '1' at the $y$-th position and ' $\mathrm{O}$ ' elsewhere.
- The cross-entropy loss will be 0 if the neural network predicts a probability of 1 for the $y$-th class and a probability of 0 for all other classes.
In summary, the cross-entropy loss is always non-negative because it measures the "distance" between the true distribution and the predicted distribution. It is 0 when the prediction matches the true distribution perfectly.

<div style="text-align: right"> <span style="color:red">[5 points]</span> </div>

Sharpness-aware minimization (SAM) (i.e., [link for main paper](https://openreview.net/pdf?id=6Tm1mposlrM) from Google Deepmind) is a simple yet but efficient technique to improve the generalization ability of deep learning models on unseen data examples. In your research or your work, you might potentially use this idea. Your task is to read the paper and implement *Sharpness-aware minimization (SAM)*. Finally, you need to apply SAM to the best architecture found in **Question 2.4**.   

```python
# creating a custom train_step to override the train_step method of sub_classing model
def train_step_sam(self, data, rho=0.05):
    """
    Overrides the train_step method of Model

    Args:
        data : Data on which model is to be trained
        rho  : Hyperparameter Rho indicating the size of neighborhood
    """

    sample_weight = None
    x, y = data

    # Opening Gradient Tape scope to record operations during 1st forward pass
    with tf.GradientTape() as tape:
        y_pred = self(x, training=True)
        # Calculating loss to calculate gradients
        loss = self.compiled_loss(y, y_pred, sample_weight=sample_weight, regularization_losses=self.losses)


    trainable_vars = self.trainable_variables
    # Calculating gradients with respect trainable variable
    gradients = tape.gradient(loss, trainable_vars)

    """
    This is the first step which involves calculating the point w_adv with highest loss and virtually moving to that point so that we can get gradient at that point.
    """
    eps_w_ls = [] # list to store the updates done to trainable variables in first step

    #computing the norm of gradients which is required for computing eps_w
    grad_norm = tf.linalg.global_norm(gradients)

    # Iterating over trainable_vars
    for i in range(len(trainable_vars)):
        # we will calculate eps_w to find w_adv point having highest loss in rho neighborhood
        eps_w = tf.math.multiply(gradients[i], rho / grad_norm )
        # temporarily moving to w_adv point
        trainable_vars[i].assign_add(eps_w)
        # storing updates done in eps_w_ls list
        eps_w_ls.append(eps_w)

    # Opening Gradient Tape scope to record operations during 2nd forward pass
    with tf.GradientTape() as tape:
        y_pred = self(x, training=True)
        # Calculating loss to calculate gradient at w_adv point
        loss = self.compiled_loss(y, y_pred, sample_weight=sample_weight, regularization_losses=self.losses)

    trainable_vars = self.trainable_variables
    #computing gradient at w_adv which is our objective in this first step
    gradients = tape.gradient(loss, trainable_vars)

    """
    This is the second step in SAM where we will do actual update at the initial point from the gradient calculated at adversial point w_adv
    """

    for i in range(len(trainable_vars)):
        # Going back to orignal parameters
        trainable_vars[i].assign_sub(eps_w_ls[i])

    # Updating parameters with gradients computed at w_adv
    self.optimizer.apply_gradients(zip(gradients, trainable_vars))

    # Updating the metrics.
    self.compiled_metrics.update_state(y, y_pred, sample_weight=sample_weight)

    # returns a dictionary mapping metric names (including the loss) to their current value.
    return {m.name: m.result() for m in self.metrics}
```

```python
import tensorflow as tf
from tensorflow.keras import layers

# Define a simple model using subclassing
class SAMModel(tf.keras.Model):
    def __init__(self, n1, n2, act):
        super(SAMModel, self).__init__()
        self.dense1 = layers.Dense(n1, activation=act)
        self.dense2 = layers.Dense(n2, activation=act)
        self.dense3 = layers.Dense(10, activation='softmax')

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.dense3(x)

    # Incorporate the custom train_step
    def train_step(self, data):
        return train_step_sam(self, data)

# Create an instance of the model
model = SAMModel(n1=40, n2=40, act='tanh')

# Compile the model
base_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=base_optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Train the model (assuming you have X_train, y_train, X_val, y_val datasets)
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_valid, y_valid))
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping

class YourModel(DefaultModel):
    def __init__(self,
                 name='network1',
                 width=32, height=32, depth=3,
                 num_classes=4,
                 is_augmentation=False,
                 activation_func='relu',
                 optimizer='adam',
                 batch_size=32,
                 num_epochs=20,
                 learning_rate=0.001,
                 verbose=True,
                 use_skip=True,
                 num_blocks=3,  # Added num_blocks as a parameter
                 mean_pool=False):  # Added mean_pool as a parameter
        super(YourModel, self).__init__(name, width, height, depth, num_classes, is_augmentation,
                                        activation_func, optimizer, batch_size, num_epochs,
                                        learning_rate, verbose)
        self.use_skip = use_skip
        self.num_blocks = num_blocks
        self.mean_pool = mean_pool
        self.use_skip = use_skip

    def build_cnn(self):
        # Start defining the network
        input_tensor = layers.Input(shape=(self.width, self.height, self.depth))
        x = input_tensor

        for _ in range(self.num_blocks):
            # Convolutional layer
            x = layers.Conv2D(16 * (_ + 1), (3, 3), strides=(1, 1), padding='SAME')(x)
            x = layers.BatchNormalization()(x)
            x = layers.Activation(self.activation_func)(x)

            skip = x

            # Convolutional layer
            x = layers.Conv2D(16 * (_ + 1), (3, 3), strides=(1, 1), padding='SAME')(x)
            x = layers.BatchNormalization()(x)
            if self.use_skip:
                x = layers.Add()([x, skip])
            x = layers.Activation(self.activation_func)(x)

            # Pooling layer
            x = layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2), padding='SAME')(x)

        # Apply Global Average Pooling or Flatten based on mean_pool flag
        if self.mean_pool:
            x = layers.GlobalAveragePooling2D()(x)
        else:
            x = layers.Flatten()(x)

        # Dense layer for classification
        output = layers.Dense(self.num_classes, activation='softmax')(x)
        self.model = models.Model(inputs=input_tensor, outputs=output)

        # Compile the model
        self.model.compile(optimizer=self.optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    def fit(self, data_manager):
        early_stopping = EarlyStopping(monitor='val_accuracy', patience=3)

        if self.is_augmentation:
            datagen = ImageDataGenerator(
                rotation_range=20,
                width_shift_range=0.2,
                height_shift_range=0.2,
                horizontal_flip=True
            )
            training_data = datagen.flow(data_manager.X_train, data_manager.y_train, batch_size=self.batch_size)

            history = self.model.fit(
                training_data,
                epochs=self.num_epochs,
                validation_data=(data_manager.X_valid, data_manager.y_valid),
                callbacks=[early_stopping],
                verbose=self.verbose
            )
        else:
            history = self.model.fit(
                data_manager.X_train, data_manager.y_train,
                batch_size=self.batch_size,
                epochs=self.num_epochs,
                validation_data=(data_manager.X_valid, data_manager.y_valid),
                callbacks=[early_stopping],
                verbose=self.verbose
            )

        return history
```

```python
import numpy as np
import tensorflow as tf

class YourModelWithMixup(YourModel):
    def __init__(self, alpha=0.2, *args, **kwargs):
        super(YourModelWithMixup, self).__init__(*args, **kwargs)
        self.alpha = alpha

    def generate_mixed_batch(self, loader1, loader2):
        for (x1, y1), (x2, _) in zip(loader1, loader2):
            lam = np.random.beta(self.alpha + 1, self.alpha, len(x1))
            x_mixed = lam.reshape(-1, 1, 1, 1) * x1 + (1. - lam.reshape(-1, 1, 1, 1)) * x2
            y_mixed = y1  # Only the y1 labels are used
            yield x_mixed, y_mixed

    def fit(self, data_manager):
        early_stopping = EarlyStopping(monitor='val_accuracy', patience=3)

        if self.is_augmentation:
            datagen1 = ImageDataGenerator(
                rotation_range=20,
                width_shift_range=0.2,
                height_shift_range=0.2,
                horizontal_flip=True
            )
            datagen2 = ImageDataGenerator(
                rotation_range=20,
                width_shift_range=0.2,
                height_shift_range=0.2,
                horizontal_flip=True
            )

            loader1 = datagen1.flow(data_manager.X_train, data_manager.y_train, batch_size=self.batch_size)
            loader2 = datagen2.flow(data_manager.X_train, data_manager.y_train, batch_size=self.batch_size)

            mixed_loader = self.generate_mixed_batch(loader1, loader2)

            steps_per_epoch = len(data_manager.X_train) // self.batch_size

            history = self.model.fit(
                mixed_loader,
                epochs=self.num_epochs,
                steps_per_epoch=steps_per_epoch,
                validation_data=(data_manager.X_valid, data_manager.y_valid),
                callbacks=[early_stopping],
                verbose=self.verbose
            )
        else:
            history = super().fit(data_manager)

        return history
```

<span style="color:red">**Question 3.5**</span> Implement the **one-versus-all (OVA)** loss. The details are as follows:
- You need to apply `the sigmoid activation function` to logits $h = [h_1, h_2,...,h_M]$ instead of `the softmax activation` function as usual to obtain $p = [p_1, p_2,...,p_M]$, meaning that $p_i = sigmoid(h_i), i=1,...,M$. Note that $M$ is the number of classes.
- Given a data example $x$ with the ground-truth label $y$, the idea is to maximize the likelihood $p_y$ and to minimize the likelihoods $p_i, i \neq y$. Therefore, the objective function is to find the model parameters to
  - $\max\left\{ \log p_{y}+\sum_{i\neq y}\log(1-p_{i})\right\}$ or equivalently $\min\left\{ -\log p_{y}-\sum_{i\neq y}\log(1-p_{i})\right\}$.
  - For example, if $M=3$ and $y=2$, you need to minimize $\min\left\{ -\log(1-p_{1})-\log p_{2}-\log(1-p_{3})\right\}$.

Compare the model trained with the OVA loss and the same model trained with the standard cross-entropy loss.

<div style="text-align: right"> <span style="color:red">[5 points]</span> </div>



```python
def ova_loss(y_true, y_pred):
    y_true_onehot = tf.one_hot(tf.cast(y_true, 'int32'), depth=y_pred.shape[-1])
    pos_loss = -y_true_onehot * tf.math.log(y_pred + tf.keras.backend.epsilon())
    neg_loss = -(1 - y_true_onehot) * tf.math.log(1 - y_pred + tf.keras.backend.epsilon())
    return tf.reduce_mean(tf.reduce_sum(pos_loss + neg_loss, axis=-1))
```

```python
class YourModel(DefaultModel):
    def __init__(self,
                 loss_type="cross_entropy",  # default loss is cross_entropy
                 name='network1',
                 width=32, height=32, depth=3,
                 num_classes=4,
                 is_augmentation=False,
                 activation_func='relu',
                 optimizer='adam',
                 batch_size=32,
                 num_epochs=20,
                 learning_rate=0.001,
                 verbose=True,
                 use_skip=True,
                 num_blocks=3,
                 mean_pool=False):
        super(YourModel, self).__init__(name, width, height, depth, num_classes, is_augmentation,
                                        activation_func, optimizer, batch_size, num_epochs,
                                        learning_rate, verbose)
        self.use_skip = use_skip
        self.num_blocks = num_blocks
        self.mean_pool = mean_pool
        self.loss_type = loss_type

    def build_cnn(self):
        input_tensor = layers.Input(shape=(self.width, self.height, self.depth))
        x = input_tensor

        for _ in range(self.num_blocks):
            # Convolutional layer
            x = layers.Conv2D(16 * (_ + 1), (3, 3), strides=(1, 1), padding='SAME')(x)
            x = layers.BatchNormalization()(x)
            x = layers.Activation(self.activation_func)(x)
            skip = x
            # Convolutional layer
            x = layers.Conv2D(16 * (_ + 1), (3, 3), strides=(1, 1), padding='SAME')(x)
            x = layers.BatchNormalization()(x)
            if self.use_skip:
                x = layers.Add()([x, skip])
            x = layers.Activation(self.activation_func)(x)
            # Pooling layer
            x = layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2), padding='SAME')(x)

        # Apply Global Average Pooling or Flatten based on mean_pool flag
        if self.mean_pool:
            x = layers.GlobalAveragePooling2D()(x)
        else:
            x = layers.Flatten()(x)

        # Change the activation function of the output layer based on loss type
        if self.loss_type == "ova":
            output = layers.Dense(self.num_classes, activation='sigmoid')(x)
        else:
            output = layers.Dense(self.num_classes, activation='softmax')(x)

        self.model = models.Model(inputs=input_tensor, outputs=output)

        # Compile the model with the appropriate loss
        if self.loss_type == "ova":
            self.model.compile(optimizer=self.optimizer, loss=ova_loss, metrics=['accuracy'])
        else:
            self.model.compile(optimizer=self.optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
def fgsm_attack(model, input_image, input_label=None,
                epsilon=0.3,
                clip_value_min=0.,
                clip_value_max=1.0,
                soft_label=False,
                from_logits=True):

    # Use categorical cross entropy
    loss_fn = tf.keras.losses.sparse_categorical_crossentropy

    # Convert numpy arrays to tensors and ensure type float32
    input_image = tf.convert_to_tensor(input_image, dtype=tf.float32)

    if input_label is not None:
        if soft_label:
            input_label = tf.convert_to_tensor(input_label, dtype=tf.float32)
        else:
            input_label = tf.convert_to_tensor(input_label, dtype=tf.int64)

    with tf.GradientTape() as tape:
        tape.watch(input_image)
        output = model(input_image)
        if not soft_label:
            loss = loss_fn(input_label, output, from_logits=from_logits)
        else:
            pred_label = tf.math.argmax(output, axis=1)
            loss = loss_fn(pred_label, output, from_logits=from_logits)

    gradient = tape.gradient(loss, input_image)
    adv_image = input_image + epsilon * tf.sign(gradient)
    adv_image = tf.clip_by_value(adv_image, clip_value_min, clip_value_max)
    adv_image = tf.stop_gradient(adv_image)

    return adv_image

def pgd_attack(model, input_image, input_label=None,
               epsilon=0.0313,
               num_steps=20,
               step_size=0.002,
               clip_value_min=0.,
               clip_value_max=1.0,
               soft_label=False,
               from_logits=True):

    loss_fn = tf.keras.losses.sparse_categorical_crossentropy

    input_image = tf.convert_to_tensor(input_image, dtype=tf.float32)

    if input_label is not None:
        if soft_label:
            input_label = tf.convert_to_tensor(input_label, dtype=tf.float32)
        else:
            input_label = tf.convert_to_tensor(input_label, dtype=tf.int64)

    random_noise = tf.random.uniform(shape=input_image.shape, minval=-epsilon, maxval=epsilon, dtype=tf.float32)
    adv_image = input_image + random_noise

    for _ in range(num_steps):
        with tf.GradientTape() as tape:
            tape.watch(adv_image)
            adv_output = model(adv_image)
            if not soft_label:
                loss = loss_fn(input_label, adv_output, from_logits=from_logits)
            else:
                pred_label = tf.math.argmax(adv_output, axis=1)
                loss = loss_fn(pred_label, adv_output, from_logits=from_logits)

        gradient = tape.gradient(loss, adv_image)
        adv_image = adv_image + step_size * tf.sign(gradient)
        adv_image = tf.clip_by_value(adv_image, input_image-epsilon, input_image+epsilon)
        adv_image = tf.clip_by_value(adv_image, clip_value_min, clip_value_max)
        adv_image = tf.stop_gradient(adv_image)

    return adv_image

def mim_attack(model, input_image, input_label=None,
               epsilon=0.3,
               gamma=0.9,
               num_steps=20,
               step_size=0.01,
               clip_value_min=0.,
               clip_value_max=1.0,
               soft_label=False,
               from_logits=True):

    loss_fn = tf.keras.losses.sparse_categorical_crossentropy

    input_image = tf.convert_to_tensor(input_image, dtype=tf.float32)

    if input_label is not None:
        if soft_label:
            input_label = tf.convert_to_tensor(input_label, dtype=tf.float32)
        else:
            input_label = tf.convert_to_tensor(input_label, dtype=tf.int64)

    random_noise = tf.random.uniform(shape=input_image.shape, minval=-epsilon, maxval=epsilon, dtype=tf.float32)
    adv_image = input_image + random_noise
    adv_noise = random_noise

    for _ in range(num_steps):
        with tf.GradientTape() as tape:
            tape.watch(adv_image)
            adv_output = model(adv_image)
            if not soft_label:
                loss = loss_fn(input_label, adv_output, from_logits=from_logits)
            else:
                pred_label = tf.math.argmax(adv_output, axis=1)
                loss = loss_fn(pred_label, adv_output, from_logits=from_logits)

        gradient = tape.gradient(loss, adv_image)
        adv_image_new = adv_image + step_size * tf.sign(gradient)
        adv_image_new = tf.clip_by_value(adv_image_new, input_image-epsilon, input_image+epsilon)
        adv_image_new = tf.clip_by_value(adv_image_new, clip_value_min, clip_value_max)
        adv_noise = gamma * adv_noise + (1-gamma) * (adv_image_new - adv_image)
        adv_image = adv_image_new
        adv_image = tf.stop_gradient(adv_image)

    adv_image = adv_image + adv_noise
    adv_image = tf.clip_by_value(adv_image, input_image-epsilon, input_image+epsilon)
    adv_image = tf.clip_by_value(adv_image, clip_value_min, clip_value_max)

    return adv_image
```
